```python
import numpy as np
import sys
import os
sys.path.append(os.path.abspath('..'))
from models import rod_model as model
from scripts import root_finder as rf
import matplotlib.pyplot as plt
```

```python
k_global = np.zeros((3,3))

def det_global_stiffness_matrix(EA1, EA2, L1, L2, rho1, rho2, A1, A2, omega):

    k_e1 = model.rod_model(EA1, L1, rho1, A1, omega)
    k_e2 = model.rod_model(EA2, L2, rho2, A2, omega)

    k_global = np.zeros((3,3), dtype=complex)
    k_global[0:2, 0:2] += k_e1
    k_global[1:3, 1:3] += k_e2

    k_global_constrained = k_global[1:, 1:]
    det_k_global_constrained = np.linalg.det(k_global_constrained)

    return det_k_global_constrained

```

```python
L1 = 2.5
L2 = 2.5
rho = 7850 
rho = 7850 
A1 = 1.13*10**-4
A2 = 1.13*10**-4
EA1 = A1*210e9
EA2 = A2*210e9

roots = rf.find_all_roots(lambda omega: det_global_stiffness_matrix(EA1, EA2, L1, L2, rho, rho, A1, A2, omega), 1e-3, 100000, step=0.1, tol=1e-3)
```
